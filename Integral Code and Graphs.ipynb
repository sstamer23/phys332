Part 2: Numerical Calculation of the Field

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.constants as constant

eps_0 = constant.epsilon_0

def Electric_Field (sigma, a, alpha, z):
    return ((sigma*a**2)/(2*eps_0)) * ((1/z**2) - ((a-z*np.cos(alpha)) /(z**2*np.sqrt((-2*a*z*
    np.cos(alpha))+z**2+a**2))))

field_list = []
z_list = []
for n in range (1, 501):
    field = Electric_Field(5, 1, 0.01745329, z=0.01*n*1)
    field_list.append(field)
    z_list.append(0.01*n*1)  
```

Part 3: Graphing/Interpreting the Results

```python
plt.plot(z_list, field_list)
plt.axvline(x=0.01*100*1, color='r')
plt.yscale('log')
plt.legend(['All Points', 'Radius a'])
plt.ylabel('log(E)')
plt.xlabel('z')
```

```python
# Z = 0 to a
new_z = np.arange(0, 5.01, 0.01)
E_field  = []
for zi in new_z:
    E_field.append(Electric_Field(5, 1, 0.01745329, z=zi))
    
plt.scatter(z_list[0:100:2], field_list[0:100:2], s=5)
plt.plot(new_z[0:100:1], E_field[0:100:1], color='tab:gray', alpha=0.25)
plt.axvline(x=0.01*100*1, color='r')
plt.yscale('log')
plt.legend(['Points', 'Function', 'Radius a'])
plt.ylabel('log(E)')
plt.xlabel('z')
```
